```python
from wordcloud import WordCloud
import pandas as pd
```

```python
f = open('Alice.txt').read()
```

## Basic Word Cloud

```python
wordcloud = WordCloud(background_color="white",width=1000, height=860, margin=2).generate(f) 
```

```python
import matplotlib.pyplot as plt
ax = plt.imshow(wordcloud)
fig = ax.figure
fig.set_size_inches(25,20) 
plt.axis("off")
plt.show()
```

## Word Cloud with shape

```python
from PIL import Image
import numpy as np
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
alice_coloring = np.array(Image.open("alice_color.png")) 
stopwords = set(STOPWORDS)
stopwords.add("said")
```

By changing the parameter of mask to plot the shape

```python
wc = WordCloud(background_color="white", max_words=2000, mask=alice_coloring,
               stopwords=stopwords, max_font_size=40, random_state=42)
wc.generate(f)
```

```python
image_colors = ImageColorGenerator(alice_coloring)
```

```python
plt.axis("off")
ax = plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
fig = ax.figure
fig.set_size_inches(25,20)                 
plt.figure()
```
